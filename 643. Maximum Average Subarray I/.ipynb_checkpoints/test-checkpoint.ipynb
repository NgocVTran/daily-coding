Given an array consisting of n integers, find the contiguous subarray of given length k that has the maximum average value. And you need to output the maximum average value.

Example 1:

Input: [1,12,-5,-6,50,3], k = 4
Output: 12.75
Explanation: Maximum average is (12-5-6+50)/4 = 51/4 = 12.75
 

Note:

1 <= k <= n <= 30,000.
Elements of the given array will be in the range [-10,000, 10,000].
 

```python
# more clean but slower
def findMaxAverage(nums, k):
    n = len(nums)
    if n<k:
        return sum(nums)/k
    
    summ = 0
    for ind in range(k):
        summ = summ + nums[ind]
        
    i = 0
    maxx = summ
    while i+k<=n:
        if i+k<n:
            summ = summ - nums[i] + nums[i+k]
            if maxx < summ:
                maxx = summ
        i = i+1
    
    return maxx/k
```

### Solution

The algorithm here is kind of variation of sliding window.
This d can be considered as the accumulated delta(i.e., nums[i+k] - nums[i] ), generated by moving windows on index i.
It use M, as the maximum d, to maintain maximum accumulated delta value.

Finally, compute (initial partial sum + M ) / k to get the result of maximum moving average.

When it comes to tradition sliding window, it would maintain a partial sum on each i

```python
def findMaxAverage(nums, k):
    M = d = 0
    for i in range(len(nums)-k):
        d += nums[i+k] - nums[i]
        if d > M: M = d
    return (sum(nums[:k])+M)/k
```

```python
findMaxAverage(nums, k)
```
